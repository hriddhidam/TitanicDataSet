#Cleaning Datas

```python
sns.boxplot(x='Pclass',y='Age',data=df)
```

```python
def avg_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(avg_age, axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.drop(['Cabin'],inplace=True,axis=1)
```
